```python
# import lib
from apyori import apriori

dataset = [
    ['Milk', 'Bread'],
    ['Milk', 'Diaper', 'Beer', 'Bread'],
    ['Milk', 'Diaper', 'Beer', 'Cola'],
    ['Diaper', 'Beer'],
    ['Milk', 'Diaper', 'Beer', 'Bread']
]

# Set minimum support, confidence, and lift thresholds
min_support = 0.2       # 60%
min_confidence = 0.7    # 70%
min_lift = 1

# Run Apriori algorithm
rules = apriori(
    transactions=dataset,
    min_support=min_support,
    min_confidence=min_confidence,
    min_lift=min_lift
)

# Convert the rules to a list for easier inspection
results = list(rules)

# Print association rules
for rule in results:
    items = [item for item in rule.items]
    print(f"Rule: {items} -> {list(rule.ordered_statistics[0].items_add)}")
    print(f"Support: {rule.support:.2f}")
    print(f"Confidence: {rule.ordered_statistics[0].confidence:.2f}")
    print(f"Lift: {rule.ordered_statistics[0].lift:.2f}")
    print("------")

```
